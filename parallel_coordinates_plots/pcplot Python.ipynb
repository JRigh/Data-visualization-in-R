### on normalized data

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Rearrange the columns to have the desired feature order
iris = iris[['petal_length', 'petal_width', 'sepal_length', 'sepal_width', 'species']]

# Normalize the numerical columns (petal length, petal width, sepal length, sepal width)
scaler = StandardScaler()
iris[['petal_length', 'petal_width', 'sepal_length', 'sepal_width']] = scaler.fit_transform(iris[['petal_length', 'petal_width', 'sepal_length', 'sepal_width']])

# Create a parallel coordinates plot with line transparency
plt.figure(figsize=(8, 5))
parallel_coordinates(iris, "species", colormap='viridis', alpha=0.8)  # Adjust alpha here

# Customize the plot
plt.title("Parallel coordinates plots (Normalized Data)")
plt.suptitle("Iris dataset", fontsize=12, fontstyle='italic', color='darkred')
plt.xlabel("Feature")
plt.ylabel("Standardized Value")  # Adjust the ylabel to reflect normalization

# Show the plot
plt.show()
```
