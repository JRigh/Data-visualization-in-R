# Sparklines plot in Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
np.random.seed(2023)

# creating dataset
year = np.linspace(start=1986, stop=2023, num=38)
v1 = np.random.poisson(4, len(year)) * np.linspace(start=0.1, stop=1, num=38)
v2 = np.random.poisson(6, len(year))
v3 = np.random.poisson(3.6, len(year)) * np.linspace(start=0.1, stop=1, num=38)
```

```python
dataset = pd.DataFrame({'year': year, 'v1': v1, 'v2': v2, 'v3': v3})
dataset
```

```python
# reshaping dataset
dataset1 = pd.melt(dataset, id_vars="year", value_vars = 'v1')
dataset2 = pd.melt(dataset, id_vars="year", value_vars = 'v2')
dataset3 = pd.melt(dataset, id_vars="year", value_vars = 'v3')
df12 = dataset1.append(dataset2)
df123 = df12.append(dataset3)
```

```python
# plotiing
sns.relplot(data = df123, y='value', x='year', row = 'variable', kind='line', height=1, aspect=4)
plt.figure(figsize=(6, 6))
plt.show()
```
